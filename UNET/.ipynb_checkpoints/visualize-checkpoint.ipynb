```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from scaler import LogScaler
import UNET
from sklearn.metrics import r2_score
from classUnet import Unet
```

```python
#保存モデル読み込み
model = Unet()
model.load_state_dict(torch.load("UNET_0.983(mape).pth"))

test_pred = UNET.scaler_out.inverse(model(UNET.twoD_x_test_tensor_scaled, UNET.oneD_x_test_tensor_scaled))
```

```python
#skip connection1回目の前
def visualize(data, n):
    fig, axes = plt.subplots(4, 4, figsize=(10, 6))
    for i, ax in zip(range(16), axes.ravel()):
        ax.imshow(data.detach().numpy()[n][i])
        ax.set_title(f"{i}", fontsize=20)
        ax.axis("off")
    fig.tight_layout()
```

```python
#最終出力
def last_visualize(n):
    fig, axes=plt.subplots(1, 2, figsize=(20, 10))
    axes[0].imshow(UNET.y_test[n, 0, :, :], vmin=0, vmax=18)
    axes[0].set_title("Sim", fontsize=30)
    axes[0].axis("off")
    axes[1].imshow(test_pred.detach().numpy()[n, 0, :, :], vmin=0, vmax=18)
    axes[1].set_title("ML", fontsize=30)
    axes[1].axis("off")
    print(UNET.oneD_x_test_tensor[n])
    plt.show()
```

```python
#横切りプロット
def horizontal_plot(n, y):
#     fig, axes=plt.subplots(1, 2, figsize=(20, 5))
    fig = plt.figure(figsize=(10, 5))
    twoD_pred_data = test_pred.detach().numpy()[n, 0, :, :]
    twoD_sim_data = UNET.y_test[n, 0, :, :]
    x = np.arange(0, twoD_pred_data.shape[1])
    plt.plot(x, twoD_sim_data[y, :], label="sim")
    plt.plot(x, twoD_pred_data[y, :], label="ML", linestyle="--")
    plt.legend()
```

```python
#縦切りプロット
def vertical_plot(n, x):
#     fig, axes=plt.subplots(1, 2, figsize=(7, 5))
    fig = plt.figure(figsize=(5, 5))
    twoD_pred_data = test_pred.detach().numpy()[n, 0, :, :]
    twoD_sim_data = UNET.y_test[n, 0, :, :]
    y = np.arange(0, twoD_pred_data.shape[0])
    plt.plot(y, twoD_sim_data[:, x], label="sim")
    plt.plot(y, twoD_pred_data[:, x], label="ML", linestyle="--")
    plt.legend()
```
